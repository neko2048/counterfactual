```python
import numpy as np
import pandas as pd
```

```python
subName = "contiPositiveAge"
data = pd.read_csv("./data/totalData{}.csv".format(subName)).iloc[:, 1:]
```

```python
data
```

```python
artData = data.copy(deep=True)
```

```python
yType = 2
t1 = artData['y{}'.format(yType)][(artData['T'] == 1)]
t0 = artData['y{}'.format(yType)][(artData['T'] == -1)]
s1 = artData['y{}'.format(yType)][(artData['Sex'] == 1)]
s0 = artData['y{}'.format(yType)][(artData['Sex'] == -1)]
o  = artData['y{}'.format(yType)][(artData['AgeTag'] == 1)]
y  = artData['y{}'.format(yType)][(artData['AgeTag'] == -1)]
t1s1o = artData['y{}'.format(yType)][(artData['T'] == 1) * (artData['Sex'] == 1) * (artData['AgeTag'] == 1)]
t1s1y = artData['y{}'.format(yType)][(artData['T'] == 1) * (artData['Sex'] == 1) * (artData['AgeTag'] == -1)]
t1s0o = artData['y{}'.format(yType)][(artData['T'] == 1) * (artData['Sex'] == -1) * (artData['AgeTag'] == 1)]
t1s0y = artData['y{}'.format(yType)][(artData['T'] == 1) * (artData['Sex'] == -1) * (artData['AgeTag'] == -1)]
t0s1o = artData['y{}'.format(yType)][(artData['T'] == -1) * (artData['Sex'] == 1) * (artData['AgeTag'] == 1)]
t0s1y = artData['y{}'.format(yType)][(artData['T'] == -1) * (artData['Sex'] == 1) * (artData['AgeTag'] == -1)]
t0s0o = artData['y{}'.format(yType)][(artData['T'] == -1) * (artData['Sex'] == -1) * (artData['AgeTag'] == 1)]
t0s0y = artData['y{}'.format(yType)][(artData['T'] == -1) * (artData['Sex'] == -1) * (artData['AgeTag'] == -1)]
print([len(x) for x in [t0s0y, t0s0o, t1s0y, t1s0o]])
print([len(x) for x in [t0s1y, t0s1o, t1s1y, t1s1o]])
```

```python
t1s1oArt = t1s1o - np.mean(t1) - np.mean(s1) - np.mean(o) # 3
t1s1yArt = t1s1y - np.mean(t1) - np.mean(s1) - np.mean(y) # 247
t1s0oArt = t1s0o - np.mean(t1) - np.mean(s0) - np.mean(o) # 3
t1s0yArt = t1s0y - np.mean(t1) - np.mean(s0) - np.mean(y) # 247
t0s1oArt = t0s1o - np.mean(t0) - np.mean(s1) - np.mean(o) # 247
t0s1yArt = t0s1y - np.mean(t0) - np.mean(s1) - np.mean(y) # 3
t0s0oArt = t0s0o - np.mean(t0) - np.mean(s0) - np.mean(o) # 247
t0s0yArt = t0s0y - np.mean(t0) - np.mean(s0) - np.mean(y) # 3
```

```python
from scipy.stats import rankdata
total = np.hstack((t1s1oArt, t1s1yArt, t1s0oArt, t1s0yArt, t0s1oArt, t0s1yArt, t0s0oArt, t0s0yArt))
rank = rankdata(total)
```

```python
t1s1oRank = rank[:3]
t1s1yRank = rank[3:250]
t1s0oRank = rank[250:253]
t1s0yRank = rank[253:500]
t0s1oRank = rank[500:747]
t0s1yRank = rank[747:750]
t0s0oRank = rank[750:993]
t0s0yRank = rank[993:]
print([len(x) for x in [t1s1oRank, t1s1yRank, t1s0oRank, t1s0yRank, t0s1oRank, t0s1yRank, t0s0oRank, t0s0yRank]])
```

```python
age = []
sex = []
treatment = []
for i in [t1s1oArt, t1s1yArt, t1s0oArt, t1s0yArt, t0s1oArt, t0s1yArt, t0s0oArt, t0s0yArt]:
    age += list(data['Age'][i.index])
    sex += list(data['Sex'][i.index])
    treatment += list(data['T'][i.index])
age = [round(x, 2) for x in age]
```

```python
dat = {'age':age,'sex':sex,'treatment':treatment,'output':rank}
df = pd.DataFrame(dat)
import statsmodels.api as sm
from statsmodels.formula.api import ols
mod = ols('output ~ age*sex*treatment', data = df).fit()
sm.stats.anova_lm(mod, typ = 2)
```
